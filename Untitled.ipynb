```python
import re
import datetime
from requests_html import HTMLSession

def process_content_type1(c, dic):
    if "入关" in c:
        return
    
    if  c.startswith("无症状感染者"):
        content_type = "Asymptomatic"
        mark_content = "无症状感染者"
    elif c.startswith("病例"):
        content_type = "ConfirmedCases"
        mark_content = "病例"
    else:
        return 
    
    if "本市闭环隔离管控人员" in c:
        control_type = "In_control"
    elif "为此前报告的本土无症状感染者" in c:
        control_type = "FromAsymptomatic"
    else:
        control_type = "Out_control"
        
    if not content_type in dic:
        dic[content_type] = {}
    if not control_type in dic[content_type]:
        dic[content_type][control_type] = {}
        
    c = c.split("，")
    
    for i in range(len(c)>>1):
        if c[2*i].startswith(mark_content):
            patients = c[2*i].split('—')
            
            if len(patients) == 2:
                s = int(patients[0][len(mark_content):])
                if patients[1].startswith(mark_content):
                    e = int(patients[1][len(mark_content):])
                else:
                    e = int(patients[1])
                num = e-s+1
            elif "、" in c[2*i]:
                num = len(c[2*i].split("、" ))
            else:
                num = 1
        
            addr = c[2*i+1][3:]
            dic[content_type][control_type][addr] = num

        else:
            break    
    return

def process_wp(c, dic):
    aeras = ['浦东新区', '黄浦区', '徐汇区', '长宁区', '静安区', '普陀区', '虹口区', '杨浦区', '闵行区', '宝山区', 
            '嘉定区', '金山区', '松江区', '青浦区', '奉贤区', '崇明区']
    
    for a in aeras:
        if c.startswith(a):
            dic[a] = []
            if not '居住于：' in c:
                continue
            locs = c.replace(',','').replace('，','').replace('。','').replace('、','').split('居住于：')[1].split('已对相关居住地落实终末消毒措施')[0].split('\n')
            if a == '普陀区':
                print(locs)
            dic[a] = []
            for loc in locs:
                if len(loc)>0:
                    dic[a].append(loc)
            return 
    return 


def add_data_from_href(href):
    session = HTMLSession()
    r = session.get(href)
    txt = r.html.xpath('//*[@id="activity-name"]',first=True).text
    
    sections = txt.split('\n\n')
    
    date = re.search(r'\d+月\d+日', sections[0]).group(0)
    date = "2022-"+date.replace("日","").replace("月","-")
    print(date)
    dt = datetime.datetime.strptime(date,'%Y-%m-%d').date()
    
    dic = {}
    for sec in sections:
        process_content_type1(sec.replace('\n',''), dic)
    
    
        all_confirmed_cases = re.search(r'累计本土确诊(\d+)例', sec)

        if all_confirmed_cases:
            dic['all_confirmed_cases'] = all_confirmed_cases.group(1)
        else:
            continue

        all_cured_cases = re.search(r'治愈出院(\d+)例', sec)
        if all_cured_cases:
            dic['all_cured_cases'] = all_cured_cases.group(1)

        all_in_treatment= re.search(r'在院治疗(\d+)例', sec)
        if all_in_treatment:
            dic['all_in_treatment'] = all_in_treatment.group(1)

        all_critical_cases = re.search(r'其中重症(\d+)例', sec)
        if all_critical_cases:
            dic['all_critical_cases'] = all_critical_cases.group(1)
        else:
            all_critical_cases = re.search(r'其中重型(\d+)例.危重型(\d+)例', sec)
            if all_critical_cases:
                dic['all_critical_cases'] = all_critical_cases.group(1)
                dic['all_extreme_critical_cases'] = all_critical_cases.group(2)

        all_in_observation = re.search(r'现有待排查的疑似病例(\d+)例', sec)
        if all_in_observation:
            dic['all_in_observation'] = all_in_observation.group(1)

        all_deaths =  re.search(r'死亡(\d+)例', sec)
        if all_deaths:
            dic['all_deaths'] = all_deaths.group(1)
        
    if not 'all_deaths' in dic:
        all_deaths =  re.search(r'累计死亡(\d+)例', txt)
        if all_deaths:
            dic['all_deaths'] = all_deaths.group(1)
    
    return dt, dic

def add_loc_from_href(href):
    session = HTMLSession()
    s = session.get(href)
    txt = s.html.xpath('//*[@id="page-content"]',first=True).text
    
    title = txt.split('\n')[0]
    
    print(title)
    
    date = re.search(r'\d+月\d+日', title).group(0)
    date = "2022-"+date.replace("日","").replace("月","-")
    print(date)
    dt = datetime.datetime.strptime(date,'%Y-%m-%d').date()
    
    all_sections = re.split('\d+年\d+月\d+日，',txt)

    dic = {}
    for sec in all_sections:
        process_wp(sec, dic)
    
    return dt, dic
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line, Page, Bar, Pie

all_addrs = ['浦东新区', '黄浦区', '徐汇区', '长宁区', '静安区', 
                      '普陀区', '虹口区', '杨浦区', '闵行区', '宝山区', 
                      '嘉定区', '金山区', '松江区', '青浦区', '奉贤区', '崇明区']

aeras_map = {'浦东新区':0, '黄浦区':1, '徐汇区':2, '长宁区':3, '静安区':4, '普陀区':5, '虹口区':6, '杨浦区':7, '闵行区':8, '宝山区':9, 
            '嘉定区':10, '金山区':11, '松江区':12, '青浦区':13, '奉贤区':14, '崇明区':15}

def draw_graph(data, addrs, content_types, control_types, title, sum=False, mark=False):
    all_date = [datetime.datetime.strptime(d,'%Y-%m-%d').date() for d in data.keys()]
    latest = max(all_date)
    dt = max(min(all_date), latest-datetime.timedelta(days=35))
    y_data = [[] for i in range(len(addrs))]
    x_data = []
    if sum:
        y_data.append([])
    while dt <= latest:
        if not str(dt) in data:
            dt += datetime.timedelta(days=1)
            continue
        data_for_one_day = data[str(dt)]
        sum_for_one_day = 0
        for i in range(len(addrs)):
            cur = 0
            for ti in content_types:
                if not ti in data_for_one_day:
                    continue
                for tj in control_types:
                    if not tj in data_for_one_day[ti]:
                        continue
                    if addrs[i] == "非浦东":
                        for key in data_for_one_day[ti][tj]:
                            if not "浦东" in key:
                                cur += data_for_one_day[ti][tj][key]
                        continue
                    if addrs[i] in data_for_one_day[ti][tj]:
                        cur += data_for_one_day[ti][tj][addrs[i]]
        
            y_data[i].append(cur)
            sum_for_one_day += cur
        
        if sum:
            y_data[-1].append(sum_for_one_day)
        
        x_data.append(str(dt))
            
        dt += datetime.timedelta(days=1)
        
    line = Line()
    line.add_xaxis(xaxis_data=x_data)
    for i in range(len(addrs)):
        if mark:
            line.add_yaxis(
                series_name=addrs[i],
                y_axis=y_data[i],
                is_smooth=True,
#                 markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_='max')]),
                markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_='max')])
            )
        else:
            line.add_yaxis(
                series_name=addrs[i],
                y_axis=y_data[i],
                is_smooth=True
            )
#         print(y_data[i])
    if sum:
        if mark:
            line.add_yaxis(
                series_name="总计",
                y_axis=y_data[-1],
                is_smooth=True,
#                 markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_='max')]),
                markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_='max')])
            )
        else:
            line.add_yaxis(
                series_name="总计",
                y_axis=y_data[-1],
                is_smooth=True
            )
    
    pl = 50-len(title)    
    line.set_global_opts(title_opts=opts.TitleOpts(title=title,pos_bottom='bottom',pos_left=str(pl)+'%'))
    #line.render(title+".html")
#     print(type(line))
    return line

name_map = {'all_confirmed_cases':'累计确诊病例','all_cured_cases':'累计治愈病例','all_in_treatment':'在院治疗病例','all_deaths':'累计死亡病例'}
def draw_culmulative(data, title, keys=['all_confirmed_cases','all_cured_cases','all_in_treatment','all_deaths']):
    all_date = [datetime.datetime.strptime(d,'%Y-%m-%d').date() for d in data.keys()]
    latest = max(all_date)
    dt = max(min(all_date), latest-datetime.timedelta(days=35))
    
    y_data = [[] for i in range(len(keys))]
    x_data = []

    while dt <= latest:
        if not str(dt) in data:
            dt += datetime.timedelta(days=1)
            continue
        data_for_one_day = data[str(dt)]
        
        for i in range(len(keys)):
            if keys[i] in data_for_one_day:
                y_data[i].append(data_for_one_day[keys[i]])
            else:
                y_data[i].append(0)
                
        x_data.append(str(dt))
        dt += datetime.timedelta(days=1)
        
    line = Line()
    line.add_xaxis(xaxis_data=x_data)
    for i in range(len(keys)):
        if False: #keys[i] == 'all_in_treatment':
            line.add_yaxis(
                series_name=name_map[keys[i]],
                y_axis=y_data[i],
                is_smooth=True,
                markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_='max')])
            )
        else:
            line.add_yaxis(
                series_name=name_map[keys[i]],
                y_axis=y_data[i],
                is_smooth=True
            )
    pl = 50-len(title)    
    line.set_global_opts(title_opts=opts.TitleOpts(title=title,pos_bottom='bottom',pos_left=str(pl)+'%'))
    return line

def draw_daily_increase_helper(x_data, y_data, title):

    bar = Bar()
    bar.add_xaxis(x_data)


    bar.add_yaxis('', y_data)
    
    pl = 50-len(title)   
    bar.set_global_opts(title_opts=opts.TitleOpts(title=title,pos_bottom='bottom',pos_left=str(pl)+'%'),
                       xaxis_opts=opts.AxisOpts(axislabel_opts={"rotate":50}))
    bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False),
                       markline_opts=opts.MarkLineOpts(
                        data=[
                            opts.MarkLineItem(                  
#                                 name="自定义线",
                                y = 10,              
                                )],
                        # 标签配置项，参考 `series_options.LabelOpts`
                        label_opts=opts.LabelOpts(),
                        linestyle_opts=opts.LineStyleOpts(width = 3,color = '#000000')))

    return bar


def draw_daily_increase(data, title, addrs=all_addrs, content_types=['Asymptomatic','ConfirmedCases'], control_types=['Out_control'], days=4):
    all_date = [datetime.datetime.strptime(d,'%Y-%m-%d').date() for d in data.keys()]
    latest = max(all_date)
    dt = max(min(all_date), latest-datetime.timedelta(days=days))
    y_data = [[] for i in range(len(addrs))]
    x_data = []
    mean_data = {}
    cur_data = {}

    while dt <= latest:
        if not str(dt) in data:
            dt += datetime.timedelta(days=1)
            continue
        data_for_one_day = data[str(dt)]
        for i in range(len(addrs)):
            cur = 0
            for ti in content_types:
                if not ti in data_for_one_day:
                    continue
                for tj in control_types:
                    if not tj in data_for_one_day[ti]:
                        continue
                    if addrs[i] in data_for_one_day[ti][tj]:
                        cur += data_for_one_day[ti][tj][addrs[i]]
            y_data[i].append(cur)
        dt += datetime.timedelta(days=1)
            
    for i in range(len(addrs)):
        mean_data[addrs[i]] = sum(y_data[i])/(days+1)
        cur_data[addrs[i]] = y_data[i][-1]
    
    mean_data = sorted(mean_data.items(),key=lambda x:x[1],reverse=False)
    cur_data = sorted(cur_data.items(), key=lambda x:x[1], reverse=False)
    
    
    
    y1, y2 = [],[]
    x1, x2 = [], []
    for i in range(len(mean_data)):
        key1, value1 = mean_data[i][0], mean_data[i][1]
        index1 = aeras_map[key1]
        x1.append(key1)
        y1.append(value1)
        
        key2, value2 = cur_data[i][0], cur_data[i][1]
        index2 = aeras_map[key2]
        x2.append(key2)
        y2.append(value2)

    
    return [draw_daily_increase_helper(x1, y1, title[0]),
             draw_daily_increase_helper(x2, y2, title[1])]
    
        
def draw_pie(data, title):
    all_date = [datetime.datetime.strptime(d,'%Y-%m-%d').date() for d in data.keys()]
    latest = max(all_date)
    
    data_for_today = data[str(latest)]
    attr, v1 = [], []
    
    all_confirmed_cases = 0
    if 'all_confirmed_cases' in  data_for_today:
        all_confirmed_cases = int(data_for_today['all_confirmed_cases'])
        
    all_cured_cases = 0
    if 'all_cured_cases' in  data_for_today:
        all_cured_cases = int(data_for_today['all_cured_cases'])
        attr.append('累计治愈')
        v1.append(all_cured_cases)
        
    
    all_critical_cases = 0
    if 'all_critical_cases' in  data_for_today:
        all_critical_cases = int(data_for_today['all_critical_cases'])
        attr.append('在院治疗(重)')
        v1.append(all_critical_cases)
        
    all_extreme_critical_cases = 0
    if 'all_extreme_critical_cases' in  data_for_today:
        all_extreme_critical_cases = int(data_for_today['all_extreme_critical_cases'])
        attr.append('在院治疗(危重)')
        v1.append(all_extreme_critical_cases)
        
    all_deaths = 0
    if 'all_deaths' in  data_for_today:
        all_deaths = int(data_for_today['all_deaths'])
        attr.append('累计死亡')
        v1.append(all_deaths)
        
    all_in_treatment = 0
    if 'all_in_treatment' in  data_for_today:
        all_in_treatment = int(data_for_today['all_in_treatment'])
        attr.append('在院治疗(轻)')
        v1.append(all_in_treatment - all_extreme_critical_cases - all_critical_cases)
    
    pie = Pie()
    pie.add('', [list(z) for z in zip(attr, v1)])
    pl = 50-len(title)    
    pie.set_global_opts(title_opts=opts.TitleOpts(title=title,pos_bottom='bottom',pos_left=str(pl)+'%'))
    return pie

daily_increase_summary = draw_graph(all_data, ['浦东新区', '非浦东'], 
          ['Asymptomatic','ConfirmedCases'],['In_control','Out_control','FromAsymptomatic'],"每日新增无症状+确诊概览", True)


daily_out_control_increase_summary = draw_graph(all_data, ['浦东新区', '非浦东'], 
          ['Asymptomatic','ConfirmedCases'],['Out_control'],"管控区外每日新增无症状+确诊概览", True)

Change_daily_increase = draw_graph(all_data, all_addrs, 
          ['Asymptomatic','ConfirmedCases'],['FromAsymptomatic'],"无症状转归",True)

all_daily_increase = draw_graph(all_data, all_addrs, 
          ['Asymptomatic','ConfirmedCases'],['In_control','Out_control','FromAsymptomatic'],"各区每日新增无症状+确诊病例")

Out_daily_increase = draw_graph(all_data, all_addrs, 
          ['Asymptomatic','ConfirmedCases'],['Out_control'],"各区管控区外每日新增无症状+确诊病例")


daily_increase_summary_only_positive = draw_graph(all_data, ['浦东新区', '非浦东'], 
          ['ConfirmedCases'],['In_control','Out_control','FromAsymptomatic'],"每日新增确诊病例概览", True, True)

daily_out_control_increase_summary_only_positive = draw_graph(all_data, ['浦东新区', '非浦东'], 
          ['ConfirmedCases'],['Out_control'],"管控区外每日新增确诊病例概览", True, True)

daily_increase_summary_only_asymptomatic = draw_graph(all_data, ['浦东新区', '非浦东'], 
          ['Asymptomatic'],['In_control','Out_control','FromAsymptomatic'],"每日新增无症状病例概览", True, True)

daily_out_control_increase_summary_only_asymptomatic = draw_graph(all_data, ['浦东新区', '非浦东'], 
          ['Asymptomatic'],['Out_control'],"管控区外每日新增无症状病例概览", True, True)

cumulative1 = draw_culmulative(all_data, title='累计病例数')
pie = draw_pie(all_data, title='本土累计病例分布')
                        
daily_increase_bar_list = draw_daily_increase(all_data, title=["五日内社会面新增均值","当日社会面新增"])

# with open("data/location.pkl",'rb') as df:
#     loc_data = pickle.load(df)

# aeras_map = { '奉贤区':0}

all_date = [datetime.datetime.strptime(d,'%Y-%m-%d').date() for d in loc_data.keys()]
latest = max(all_date)
earliest = min(all_date)

locsets = []
for k in range(len(aeras_map)+1):
    locsets.append([set() for i in range(5)])

sd = earliest
while not sd == latest:
    if not str(sd) in loc_data:
        sd = sd + datetime.timedelta(days=1)
        continue
    
    data_for_one_day = loc_data[str(sd)]
    duration = (latest - sd).days

    for area in aeras_map:
        if area in data_for_one_day:
            if duration > 14:
                    locsets[aeras_map[area]][-1] = locsets[aeras_map[area]][-1] | set(data_for_one_day[area])
            elif duration > 7:
                locsets[aeras_map[area]][-2] = locsets[aeras_map[area]][-2] | set(data_for_one_day[area])
            elif duration > 4:
                locsets[aeras_map[area]][-3] = locsets[aeras_map[area]][-3] | set(data_for_one_day[area])
            else:
                locsets[aeras_map[area]][-4] = locsets[aeras_map[area]][-4] | set(data_for_one_day[area])
    
    sd = sd + datetime.timedelta(days=1)
    
for area in aeras_map:
    for i in range(1,5):
        for j in range(i):
            locsets[aeras_map[area]][i] = locsets[aeras_map[area]][i] - locsets[aeras_map[area]][j]

            
    
loc_latest = loc_data[str(latest)]
total_list = [0 for i in range(5)]
yesterday_locs = 0

res = [[0.0 for i in range(len(aeras_map)+1)] for j in range(5)]

for area in aeras_map:
    old_places = 0.0
    
    if area in loc_latest:
        count = len(loc_latest[area])
        if count == 0:
            continue
        yesterday_locs += count
        for i in range(1,5):
            tmp = set(loc_latest[area]) & locsets[aeras_map[area]][i]
            res[i][aeras_map[area]] = 1.0* len(tmp) / count
            old_places += res[i][aeras_map[area]]
            total_list[i] += len(tmp)
            
    res[0][aeras_map[area]] = 1 - old_places
    
if yesterday_locs>0:
    for i in range(1,5):
        res[i][-1] = 1.0*total_list[i]/yesterday_locs
    res[0][-1] = 1.0*(yesterday_locs-sum(total_list))/yesterday_locs

# for i in range(len(res)):
#     print(i, res[i])
    
bar = Bar()
xaxis = list(aeras_map.keys())
xaxis.append('总计')
bar.add_xaxis(xaxis)

ytitles = ['新增阳性小区','四日有新增','四日无新增','七日无新增','十四日无新增']

for i in range(5):
    bar.add_yaxis(ytitles[i], res[i], stack='stack1')

bar_title = str(latest)+"阳性小区分布"
pl = 50-len(bar_title)   
bar.set_global_opts(title_opts=opts.TitleOpts(title=bar_title,pos_bottom='bottom',pos_left=str(pl)+'%'),
                    xaxis_opts=opts.AxisOpts(axislabel_opts={"rotate":50}),
                    yaxis_opts=opts.AxisOpts(max_=1))
bar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))

page = Page(layout=Page.SimplePageLayout)
page.add(daily_increase_summary, daily_out_control_increase_summary, all_daily_increase, Out_daily_increase,cumulative1,pie)
for b in daily_increase_bar_list:
    page.add(b)
page.add(Change_daily_increase,bar, daily_increase_summary_only_asymptomatic,daily_out_control_increase_summary_only_asymptomatic,daily_increase_summary_only_positive,daily_out_control_increase_summary_only_positive)
page.render("index.html")
```
